```python
def anagram_checker(word_a, word_b, is_case_sensitive):
  # Function: anagram_checker 
  # Propouse: Given two valid strings, check to see if they are anagrams of each other by checking for case sentivity 
  # input:    word_a and word_b, both strings 
  #           is_case_sensitive, boolean
  # output:   valid, Boolean, True if the inputs are anagrams, otherwise False

  if len(word_a) == len(word_b):  # to be an anagram they should have the same len
#trasforming the inputs to valid strings and lowercase the words to 
    if is_case_sensitive !=True : 
       word_a_char=sorted(str(word_a).lower())
       word_b_char=sorted(str(word_b).lower())
    else: 
      word_a_char=sorted(str(word_a))
      word_b_char=sorted(str(word_b))

    valid= (word_a_char==word_b_char)
  else:
    valid=False

  return valid
```
